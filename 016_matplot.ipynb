```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
arr = np.random.randint(20000, 40000, 100)
arr
```

```python
#lineplot
plt.plot(arr)
plt.plot(sorted(arr))
plt.show()
```

```python
x = [1, 3, 5, 7]
y = [2, 40, 16, 81]

plt.plot(x, y, marker = "o", ms = 10, mfc = "red", mec = "green", color = "black", lw = 2, ls = "-.", alpha = 0.5, label = "lineplot")
# ms --> marker size
# mfc --> marker face color
# mec --> marker edge color

plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#scatterplot
plt.scatter(x,y, marker = "s")
plt.show()
```

```python
x = ["apple", "banana", "kivi", "cherry"]
y = [76, 57, 34, 41]
colours = ["red", "yellow", "green", "blue"]
hatch_patterns = ["//","\\\\", "xx", "oo"]
# plt.bar(x, y, color = colours, hatch = hatch_patterns, label = x, width = 0.3, edgecolor = "black", 
#         alpha = 0.5, linewidth = 3)
plt.barh(x, y)
plt.legend()
plt.grid()
plt.show()
```

```python
# histogram ---> distribution of data in interval(bins)

arr = np.random.randint(20000, 40000, 100)
plt.hist(arr, bins = 20)
plt.show()
```

```python
x = ["apple", "banana", "kivi", "cherry"]
y = [76, 57, 34, 41]
explodes = (0.05, 0.05, 0.05, 0.05)
plt.pie(y, labels = x, autopct = "%1.f%%", shadow = True, explode = explodes)
plt.savefig("pie_chart.jpg")
plt.show()
```

```python
# sub plot
import warnings
warnings.filterwarnings("ignore")

x = np.linspace(0, 10, 100)
plt.figure(figsize = (10, 6))
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.title("sin")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("cos")

plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))
plt.title("tan")

plt.subplot(2, 2, 4)
plt.plot(x, np.log(x))
plt.title("log")
```
